# Data Preprocessing

```python
import re
from nltk.tokenize import word_tokenize     #Natural Language Toolkit bir doƒüal dil i≈üleme k√ºt√ºphanesidir
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
```

```python
def data_processing(text):
    text = text.lower()     #Metni k√º√ß√ºk harfe d√∂n√º≈üt√ºrme
    text = re.sub(r'https\S+|www\S+https\S+','',text,flags=re.MULTILINE)   #URL kaldƒ±rƒ±lƒ±yor
    text = re.sub(r'\@w+|\#','',text)         
    text = re.sub(r'[^\w\s]','',text)         #Hashtag'leri kaldƒ±rma
    text_tokens = word_tokenize(text)         
    #Getting tokens
    #Bir c√ºmleyi kelimelere b√∂lmek i√ßin word_tokenize() y√∂ntemini kullanƒ±rƒ±z .
    #Kelime tokenizasyonunun √ßƒ±ktƒ±sƒ±, makine √∂ƒürenimi uygulamalarƒ±nda metnin daha iyi anla≈üƒ±lmasƒ± i√ßin
    #Veri √áer√ßevesine d√∂n√º≈üt√ºr√ºlebilir.
    filtered_text = [w for w in text_tokens if not w in stop_words]
    return " ".join(filtered_text)
```

```python
# Veri ƒ∞≈üleme fonksiyonunun uygulanmasƒ±
text_df.text = text_df["text"].apply(data_processing)
```

```python
# Varsa Kopyalarƒ±n Kaldƒ±rƒ±lmasƒ±
text_df = text_df.drop_duplicates('text')
```

```python
from nltk.stem import PorterStemmer
```

```python
#Stemming ger√ßekle≈ütirilmesi
stemmer = PorterStemmer()
def stemming(data):
    text = [stemmer.stem(word) for word in data]
    return data
```

```python
text_df["text"] = text_df["text"].apply(lambda x: stemming(x))
```

```python
from textblob import TextBlob
```

```python
def polarity(text):
    return TextBlob(text).sentiment.polarity
```

```python
text_df["polarity"] = text_df["text"].apply(polarity)
text_df.head(10)
```

```python
# Veri √ßer√ßevesine Duyarlƒ±lƒ±k ekleme
def sentiment(label):
    if label <0:
        return "Negative"
    elif label ==0:
        return "Neutral"
    elif label>0:
        return "Positive"
```

```python
text_df['sentiment'] = text_df['polarity'].apply(sentiment)
text_df.head(10)
```

# Model Olu≈üturma

```python
from sklearn.feature_extraction.text import CountVectorizer
```

Vekt√∂rle≈ütirme: Eƒüitim i≈üleminden √∂nceki en √∂nemli adƒ±mdƒ±r. Makine √∂ƒürenmesi algoritmalarƒ± sayƒ±sal deƒüerlere ihtiya√ß duyarlar. Metin verilerinin sayƒ±salla≈ütƒ±rƒ±lmasƒ± i√ßin bir√ßok yakla≈üƒ±m kullanƒ±labilmektedir. Yakla≈üƒ±mlarƒ±n ortak √∂zelliƒüi metin verilerinin sayƒ±salla≈ütƒ±rƒ±larak temsil edilmesi ve √∂ƒürenme modeline giri≈ü olu≈üturacak √∂zelliklerin se√ßiminin saƒülanmasƒ±dƒ±r.

```python
# Bigram modeli olu≈üturmak i√ßin vekt√∂rle≈ütirme yapƒ±lmasƒ±
vect = CountVectorizer(ngram_range=(1,2)).fit(text_df['text'])
```

N-gram: Veri √ºzerinde arama ve kar≈üƒ±la≈ütƒ±rma yaparken tekrar sayƒ±sƒ±nƒ± belirlemek i√ßin kullanƒ±lan bir algoritmadƒ±r. N tekrar derecesini ifade etmektedir. N sayƒ±sƒ±na g√∂re 1:unigram, 2:bigram, 3:trigram olarak isimlendirilmektedir. Arama motorlarƒ±nda olduƒüu gibi metin arama ve birbirleri ile en √ßok ili≈ükili kelimelerin bulunmasƒ±nda kullanƒ±lmaktadƒ±r. D√∂k√ºman aramada en sƒ±k kullanƒ±lan algoritmalardan bir tanesidir, bu √∂zelliklerinin yanƒ± sƒ±ra diƒüer bazƒ± vekt√∂rle≈ütirme algoritmalarƒ±nda da kullanƒ±lƒ±rlar

```python
from sklearn.model_selection import train_test_split
```

```python
#Baƒüƒ±msƒ±z ve Baƒüƒ±mlƒ± Deƒüi≈ükenleri ayƒ±rmak ve X verilerini d√∂n√º≈üt√ºrmek
X = text_df['text']
Y = text_df['sentiment']
X = vect.transform(X)
```

```python
# Verileri %20 testle b√∂lme
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

# Logistic Regresyon

Lojistik Regresyon Analizinin kullanƒ±m amacƒ±, istatistikte kullanƒ±lan diƒüer model yapƒ±landƒ±rma teknikleri ile aynƒ±dƒ±r. En az deƒüi≈ükeni kullanarak en iyi uyuma sahip olacak ≈üekilde baƒüƒ±mlƒ± ile baƒüƒ±msƒ±z deƒüi≈ükenler arasƒ±ndaki ili≈ükiyi tanƒ±mlayabilen ve biyolojik olarak kabul edilebilir bir model kurmaktƒ±r.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
```

```python
logreg = LogisticRegression()
logreg.fit(x_train, y_train)
logreg_pred = logreg.predict(x_test)
logreg_acc = accuracy_score(logreg_pred, y_test)
print("Test doƒüruluƒüu(accuracy) : {:.2f}%".format(logreg_acc*100))
```

```python
print(classification_report(y_test, logreg_pred))
```
